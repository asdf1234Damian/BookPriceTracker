```python
import pandas as pd 
df = pd.read_csv('./books.csv')
print(df.head())
```

```python
cols = ['title']
cols += df['date'].unique().tolist()
newFormat = pd.DataFrame(columns=cols)
newFormat['title'] = df['title'].unique()
newFormat.head()
```

```python
# newFormat = newFormat.set_index('title')
for index,row in df.iterrows():
    newFormat.loc[row['title']][row['date']] = row['price']
```

```python
newFormat.head()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

cols = [x for x in newFormat.columns.values]
plt.figure(figsize=(20,10))
for index,row in newFormat.iterrows():
    plt.plot(cols,[float(x)for x in row[cols]])
plt.xticks(cols,rotation = 'vertical')
plt.show()
```
